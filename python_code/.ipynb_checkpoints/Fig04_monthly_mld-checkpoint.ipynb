```python
heat_balance_files_weak = glob.glob('/projects2/rsmas/ikamenkovich/ROAM/GISELE/roam/ocn/new_diagnostics/monthly_data/monthly1_ml_heat_balance/' 
                                + case_name + '/monthly_ml_heat_balance_2016-*_' + case_name + 'Eddies.nc')
heat_balance_files_weak = sorted(heat_balance_files_weak)
heat_balance_files_weak
```

```python
weak_ds = xr.open_mfdataset(heat_balance_files_weak, concat_dim = 'ocean_time',combine='nested')
```

```python
import calendar
```

```python
mon = [1, 5, 8, 11]
month_name = ['Feb.', 'June', 'Sept.', 'Dec.']
vmax = [20, 200, 450, 40]
```

```python
figure = plt.figure(figsize = (7, 7))

i = 0
for n, month in enumerate(mon):
    plt.subplot(4, 2, i + 1)
    plt.pcolormesh(-1 * strong_ds.MLD[month, 60:350, 60:1062], cmap = cmo.cm.deep, vmin = 0, vmax = vmax[n])
    plt.colorbar()
    #plt.contour(-1 * strong_ds.MLD[month, 60:350, 60:1062])
    plt.title('2016-' + str(month + 1).zfill(2))
    plt.xlabel('Distance (km)')
    plt.ylabel('Distance (km)')
    i += 2
    print(i, month)

i = 1     
for n, month in enumerate(mon):
    print(i, month)
    plt.subplot(4, 2, i + 1)
    plt.pcolormesh(-1 * weak_ds.MLD[month, 60:350, 60:1062], cmap = cmo.cm.deep, vmin = 0, vmax = vmax[n])
    plt.colorbar()
    #plt.contour(-1 * strong_ds.MLD[month, 60:350, 60:1062])
    plt.title('2016-' + str(month + 1).zfill(2))
    plt.xlabel('Distance (km)')
    plt.ylabel('Distance (km)')
    i += 2


plt.tight_layout()
plt.savefig('./MLD.png')
```

```python
figure = plt.figure(figsize = (6, 6))
plt.subplot(211)
#plt.plot(np.arange(1, 13), -1 * strong_ds.MLD[:, 60:350, 60:1062].mean(axis = (-1, -2)), color = 'red')
plt.errorbar(np.arange(1, 13), -1 * strong_ds.MLD[:, 60:350, 60:1062].mean(axis = (-1, -2)), yerr = -1 * strong_ds.MLD[:, 60:350, 60:1062].std(axis = (-1, -2)), color = 'red')
plt.xlim(.9, 12.5)
plt.ylim(0, 450)
plt.title('a) SC, monthly average and standard deviation of MLD')
plt.ylabel('meters')
plt.xticks(range(1, 13))

plt.subplot(212)
#plt.plot(np.arange(1, 13), -1 * weak_ds.MLD[:, 60:350, 60:1062].mean(axis = (-1, -2)), color = 'green')
plt.errorbar(np.arange(1, 13), -1 * weak_ds.MLD[:, 60:350, 60:1062].mean(axis = (-1, -2)), yerr = -1 * weak_ds.MLD[:, 60:350, 60:1062].std(axis = (-1, -2)), color = 'green')
plt.xlim(.9, 12.5)
plt.ylim(0, 450)
plt.title('b) WC, monthly average and standard deviation of MLD')
plt.xlabel('Month')
plt.ylabel('meters')
plt.xticks(range(1, 13))

plt.tight_layout()
plt.savefig('./MLD_time_series.pdf')
```
