```python
def eddy_filter_300km(var):
    
    from scipy.ndimage import uniform_filter
    '''
    Input: varibale numpy arary with dimensions [time, z, y, z]
    Return: 500km-rolling-mean var and eddy fields
    '''
    import numpy as np
    
    var_500km_mean, var_eddy = np.ma.masked_all(var.shape), np.ma.masked_all(var.shape)
        
    for t in range(var.shape[0]):
        for s_rho in range(var.shape[1]):
            var_500km_mean[t, s_rho] = uniform_filter(var[t, s_rho], size = 35, mode = 'constant')
            var_eddy[t, s_rho]       = var[t, s_rho] - var_500km_mean[t, s_rho]
            
    return var_eddy
```

# Wind stress

```python
TauX2_WC_mean  = WIND_WC_ds.TauX2.data
TauY2_WC_mean  = WIND_WC_ds.TauY2.data
TauX2_SC_mean  = WIND_SC_ds.TauX2.data
TauY2_SC_mean  = WIND_SC_ds.TauY2.data
```

# SST gradient

```python
SST2_SC_mean_grady = np.gradient(SST2_SC, axis = 1) / 9000 # 9km horizontal resolution
SST2_SC_mean_gradx = np.gradient(SST2_SC, axis = 2) / 9000
SST2_WC_mean_grady = np.gradient(SST2_WC, axis = 1) / 9000
SST2_WC_mean_gradx = np.gradient(SST2_WC, axis = 2) / 9000
```

# Ekman heat advection

```python
f = -0.00010300000000000002
```

```python
EHT_SC =  (SST2_SC_mean_grady * TauX2_SC_mean - SST2_SC_mean_gradx * TauY2_SC_mean ) * 3850 / f 
EHT_WC =  (SST2_WC_mean_grady * TauX2_WC_mean - SST2_WC_mean_gradx * TauY2_WC_mean ) * 3850 / f 
```
