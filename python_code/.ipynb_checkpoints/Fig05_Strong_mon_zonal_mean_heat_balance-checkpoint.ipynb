# Monthly zonal mean heat balance

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean as cmo
import glob
from tqdm import tqdm
import sys
import calendar
```

```python
case_name = 'Strong'
```

```python
heat_balance_files = glob.glob('../monthly_ml_heat_budget_data_small_mesoscale/' + case_name + '/monthly_ml_heat_balance_2016-*_' + case_name + 'Eddies.nc')
heat_balance_files = sorted(heat_balance_files)
len(heat_balance_files), heat_balance_files[:]
```

# Monthly ML heat advection and surface heat flux

```python
plot_months = [2, 6, 9, 11, 12]
len(plot_months)
```

```python
fig, axs = plt.subplots(nrows = 3, ncols = 2, figsize=(6, 6))
axs = axs.ravel()
axs[-1].axis('off')

count = 0 
#for num, heat_balance_file in enumerate(useful_files[4::3]):
for num, month in enumerate(plot_months):  
    
    print(heat_balance_files[month - 1][-23:-16])
    
    ds = xr.open_dataset(heat_balance_files[month - 1])
    
    temp_rate = ds.temp_rate_ml.data[:,60:1062].mean(axis = -1)
    temp_adv  = ds.temp_adv_ml.data[:,60:1062].mean(axis = -1) #.mean(axis = 0)
    shflux = ds.shflux_ml.data[:,60:1062].mean(axis = -1) 
    res = ds.residual.data[:,60:1062].mean(axis = -1) 
    
    y_array = ds.y_rho.data[:, 0]
    
    l1 = axs[num].plot(y_array/1000, temp_rate, linewidth = 3,color = 'black')
    l2 = axs[num].plot(y_array/1000, temp_adv, linewidth = 3, color = 'red')
    l5 = axs[num].plot(y_array/1000, shflux, linewidth = 3, color = 'blue')
    l6 = axs[num].plot(y_array/1000, res, linewidth = 3, color = 'gray')
    axs[num].set_xlim([60, 1025])
    axs[num].set_ylim([-350, 150])
    axs[num].set_xlabel('N-S distance(km)', fontsize = 12 )
    axs[num].set_ylabel('$W/m^2$', fontsize = 12 )
    axs[num].set_title(heat_balance_files[month-1][-23:-16], fontsize = 15 )
    count += 1
    plt.tight_layout()

line_labels = ['ML heating/cooling', 'Total heat advection' ,'Diabatic heating', 'Residual']
fig.legend([l1, l2, l5, l6],     # The line objects
           labels=line_labels,   # The labels for each line
           loc="lower right",   # Position of legend
           borderaxespad = 2.3,    # spacing around legend box
           title=" ", # Title for the legend
           prop={'size': 12})
plt.savefig('./Fig05_SC_zonal_mean_budget.pdf')
```
