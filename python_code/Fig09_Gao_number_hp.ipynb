# Scale analysis: the relative importance of the oceanic heat advection and diabatic heating

```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean as cmo
from matplotlib import ticker, cm
```

```python
Q = 10
rho = 1030
Cp = 3850
delta_T = .5
L = 60000
```

```python
time_scale = np.arange(1e5, .3e7, .3e5)
mld = np.arange(1, 350, 3.5)
```

```python
num = np.zeros([97, 100])
for time in range(97):
    for h in range(100):
        #num[time, h] =  Q * time_scale[time]/ (rho * Cp * delta_T * mld[h]) #G
        num[time, h] =  (rho * Cp * delta_T * mld[h]) / (Q * time_scale[time]) #adv/shflux
```

```python
mld_sc2, u_sc2 = 11.57, 0.048 # m, cm/s 2016-03

mld_sc6, u_sc6 = 132, 0.048 # m, cm/s 2016-06

mld_sc9, u_sc9 = 311, 0.046 # m, cm/s 2016-09

mld_sc12, u_sc12 = 11.73, 0.056 # m, cm/s 2016-12
```

```python
mld_wc2, u_wc2 = 18, 0.027 # m, cm/s 2016-03

mld_wc6, u_wc6 = 103,  0.025# m, cm/s 2016-06

mld_wc9, u_wc9 = 333, 0.025 # m, cm/s 2016-09

mld_wc12, u_wc12 = 38, 0.029 # m, cm/s 2016-12
```

```python
std_sc9, std_sc6, std_sc2, std_sc12 = 47, 35, 5, 5
std_wc9, std_wc6, std_wc2, std_wc12 = 94, 43, 9, 17
```

```python
tau_sc9, tau_sc6, tau_sc2, tau_sc12 = L/u_sc9/86400, L/u_sc6/86400, L/u_sc2/86400, L/u_sc12/86400
tau_wc9, tau_wc6, tau_wc2, tau_wc12 = L/u_wc9/86400, L/u_wc6/86400, L/u_wc2/86400, L/u_wc12/86400
```

```python
x_sc2 = np.arange(mld_sc2 - std_sc2, mld_sc2 + std_sc2)
x_sc6 = np.arange(mld_sc6 - std_sc6, mld_sc6 + std_sc6)
x_sc9 = np.arange(mld_sc9 - std_sc9, mld_sc9 + std_sc9)
x_sc12 = np.arange(mld_sc12 - std_sc12, mld_sc12 + std_sc12)
```

```python
x_wc2 = np.arange(mld_wc2 - std_wc2, mld_wc2 + std_wc2)
x_wc6 = np.arange(mld_wc6 - std_wc6, mld_wc6 + std_wc6)
x_wc9 = np.arange(mld_wc9 - std_wc9, mld_wc9 + std_wc9)
x_wc12 = np.arange(mld_wc12 - std_wc12, mld_wc12 + std_wc12)
```

```python
fig, ax = plt.subplots(figsize = (7, 5))
cs = ax.contourf(mld, time_scale/86400, num, levels = [1, 2, 5, 8, 10, 20, 30, 100], locator = ticker.LogLocator(), cmap = cm.PuBu_r)
cs1 = ax.contour(mld, time_scale/86400, num, levels = [1, 2, 5, 8, 10, 20, 30, 100], colors = 'black')
label1 = ax.clabel(cs1, cs1.levels[:], fmt = '%1.1f',inline=True,  fontsize=12)

#cs2 = ax.plot(x_sc9, np.repeat(tau_sc9, x_sc9.size), 'r--')
#cs3 = ax.plot(x_sc2, np.repeat(tau_sc2, x_sc2.size), 'r--')
#cs6 = ax.plot(x_sc6, np.repeat(tau_sc6, x_sc6.size), 'r--')
#cs3 = ax.plot(x_sc12, np.repeat(tau_sc12, x_sc12.size), 'r--')
#cs4 = ax.plot(x_wc9, np.repeat(tau_wc9, x_wc9.size), 'g--')
#cs5 = ax.plot(x_wc2, np.repeat(tau_wc2, x_wc2.size), 'g--')
#cs7 = ax.plot(x_wc6, np.repeat(tau_wc6, x_wc6.size), 'g--')
#cs5 = ax.plot(x_wc12, np.repeat(tau_wc12, x_wc12.size), 'g--')

ax.text(x_sc12[0], tau_sc12 - 2, 'Dec.', fontsize = 14, color = 'red')
ax.text(x_sc9[20], tau_sc9 + 1, 'Sept.', fontsize = 14, color = 'red')
ax.text(x_sc6[20], tau_sc6 - 2, 'June', fontsize = 14, color = 'red')
ax.text(x_sc2[0], tau_sc2 + 1, 'Feb.', fontsize = 14, color = 'red')

ax.text(x_wc12[0], tau_wc12 - 2, 'Dec.', fontsize = 14, color = 'k')
ax.text(x_wc9[70], tau_wc9 + 1, 'Sept.', fontsize = 14, color = 'k')
ax.text(x_wc2[0], tau_wc2 + 1, 'Feb.', fontsize = 14, color = 'k')
ax.text(x_wc6[20], tau_wc6 + 1, 'June', fontsize = 14, color = 'k')


cbar = fig.colorbar(cs)
#plt.ylim([10, 30])
plt.xlabel('mld (m)', fontsize = 16)
plt.ylabel('Advective time scale (days)', fontsize = 16)

marker_size = 9

m1 = plt.plot(mld_sc12, tau_sc12, marker='x', markersize = marker_size, color="r")
m2 = plt.plot(mld_sc9, tau_sc9, marker='*', markersize = marker_size, color="r")
m3 = plt.plot(mld_sc6, tau_sc6, marker='v', markersize = marker_size, color="r")
m4 = plt.plot(mld_sc2, tau_sc2, marker='s', markersize = marker_size, color="r")

m5 = plt.plot(mld_wc12, tau_wc12, marker='x', markersize = marker_size, color="k")
m6 = plt.plot(mld_wc9, tau_wc9, marker='*', markersize = marker_size, color="k")
m7 = plt.plot(mld_wc6, tau_wc6, marker='v', markersize = marker_size, color="k")
m8 = plt.plot(mld_wc2, tau_wc2, marker='s', markersize = marker_size, color="k")

#plt.title('Gao number', fontsize = 16)
plt.savefig('Fig09_gao_number_hp.pdf', dpi = 100)
```
