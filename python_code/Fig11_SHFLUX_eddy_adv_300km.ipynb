```python
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cmocean as cmo
import glob
from tqdm import tqdm
import sys
import calendar
```

```python
def eddy_filter_300km(var):
    
    from scipy.ndimage import uniform_filter
    '''
    Input: varibale numpy arary with dimensions [z, y, z]
    Return: 500km-rolling-mean var and eddy fields
    '''
    import numpy as np
    
    var_500km_mean, var_eddy = np.ma.masked_all(var.shape), np.ma.masked_all(var.shape)
        
    for s_rho in range(var.shape[0]):
        var_500km_mean[s_rho] = uniform_filter(var[s_rho], size = 120, mode = 'constant')
        var_eddy[s_rho]       = var[s_rho] - var_500km_mean[s_rho]
            
    return var_eddy
```

```python
x_array = np.arange(0, 1122*2500, 2500)
y_array = np.arange(20*2500, 430*2500, 2500)
x_rho, y_rho = np.meshgrid(x_array, y_array)
```

```python
heat_balance_ds_strong = xr.open_mfdataset(heat_balance_files[:], concat_dim = 'ocean_time', combine = 'nested')
heat_balance_ds_strong
```

```python
shflux_strong = heat_balance_ds_strong.shflux_ml.data[:].compute()
ADV_EDDIES_strong = heat_balance_ds_strong.eddy_adv_ml.data[:, 60:350, 60:1062].compute()
shflux_eddies_strong = eddy_filter_300km(shflux_strong)
SSTA_strong = heat_balance_ds_strong.SSTA.data.compute()
```
